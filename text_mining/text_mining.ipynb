```python
import pandas as pd
import numpy as np
from hazm import word_tokenize
import hazm
```

```python
stemmer = hazm.Stemmer()
```

```python
data=pd.read_csv('per.csv', encoding='utf8')
data
```

```python
with open('stopwords.txt','r', encoding='utf8') as stopword_file:
    stopwords=stopword_file.readlines()
stopwords=[letter.replace('\n', '') for letter in stopwords]
stopwords
```

```python
df=pd.DataFrame(columns=['title_body', 'category'])
df
```

```python
for index, row in data.iterrows():
    title_body=row['Title']+ ' ' +row['Body']
    title_body_tokenize=word_tokenize(title_body)
    title_body_tokenize_filtered=[word  for word in title_body_tokenize if word not in stopwords]
    title_body_tokenize_filtered_stemmed=[stemmer.stem(word) for word in title_body_tokenize_filtered]
    df.loc[index] = {
        'title_body': ' '.join(title_body_tokenize_filtered_stemmed),
        'category': row['Category2'].replace('\n', '')
    }   
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorize=TfidfVectorizer()
vectorize.fit(df['title_body'])
```

```python
X=vectorize.transform(df['title_body'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
y=le.fit_transform(df['category'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
 rf=RandomForestClassifier(random_state=123)
rf.fit(X_train, y_train)
```

```python
rf.score(X_test, y_test)
```
