```python
df.drop(['adult_male','deck','alive'],axis=1,  inplace=True)
```

```python
df['age']=df['age'].fillna(df['age'].mean())
```

```python
df.dropna(subset=['embarked'], inplace=True)
```

```python
df.dropna(subset=['embark_town'], inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
encoder = OrdinalEncoder()
encoder.fit(df[['class']])
df['class'] = encoder.transform(df[['class']])
```

```python
df
```

```python
X=df.drop('survived', axis=1)
```

```python
y=df['survived']
```

```python
X
```

```python
ct = ColumnTransformer([('ll', OneHotEncoder(drop='first'), [1,6,8,9,10])], remainder='passthrough')
X=ct.fit_transform(X)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(X_train, y_train)
```

```python
pred_lr=LR.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred_lr))
```

```python
print(classification_report(y_test, pred_lr))
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn=KNeighborsClassifier(n_neighbors=5)
kn.fit(X_train, y_train)
```

```python
pred_kn=kn.predict(X_test)
```

```python
print(classification_report(y_test, pred_kn))
```

```python
print(classification_report(y_test, pred_kn))
```

```python
from sklearn.svm import SVC
```

```python
sv=SVC(kernel='rbf')
sv.fit(X_train, y_train)
```

```python
pred_sv=sv.predict(X_test)
```

```python
print(classification_report(y_test, pred_sv))
```

```python
print(classification_report(y_test, pred_sv))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=1000, max_depth=3)
rf.fit(X_train, y_train)
```

```python
pred_rf=rf.predict(X_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
pipe = Pipeline([('scaler', StandardScaler()), ('lr',LogisticRegression())])
```

```python
pipe.fit(X_train, y_train)
```

```python
pipe.predict(X_test)
```

```python
pipe.score(X_test, y_test)
```
