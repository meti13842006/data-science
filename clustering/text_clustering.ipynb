```python
import pandas as pd
import numpy as np
import os
import zipfile
```

```python
zip_name='MedicalText.zip'
with zipfile.ZipFile(zip_name, 'r') as zip_ref:
    file_name=zip_ref.namelist()[0]
    zip_ref.extractall()
os.remove(zip_name)
```

```python
df=pd.read_csv(file_name, delimiter='\\t', engine='python', skiprows=9000)
df.head()
```

```python
df.info()
```

```python
df.columns[0]
```

```python
new_column_names = {df.columns[0]: 'labels', df.columns[-1]: 'condition'}
df.rename(columns=new_column_names, inplace=True)
```

```python
X=df[['condition']]
```

```python
X.head()
```

```python
with open('stopwords.txt', encoding='utf8') as stopwords_file:
    stopwords = stopwords_file.readlines()
stopwords = [line.replace('\n', '') for line in stopwords]
stopwords
```

```python
from nltk import word_tokenize
import nltk
import re
nltk.download('punkt')
dataset = pd.DataFrame(columns=['title_body'])
for index, row in X.iterrows():
    title_body_tokenized = word_tokenize(row['condition'])
    title_body_tokenized_filtered = [w.lower() for w in title_body_tokenized if not w.lower() in stopwords]
    #     title_body_tokenized_filtered_stemmed = [stemmer.stem(w) for w in title_body_tokenized_filtered]
    s = re.sub('[^\w\s]', '', ' '.join(title_body_tokenized_filtered))
    s = re.sub("\d+", "", s)
    dataset.loc[index] = {
        'title_body': s,
    }
```

```python
dataset
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(dataset['title_body'])
```

```python
X = vectorizer.transform(dataset['title_body'])
```

```python
X
```

```python
from sklearn.cluster import SpectralClustering
```

```python
msh=SpectralClustering()
msh.fit(X)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, msh.labels_)
```

```python
np.unique(msh.labels_)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300)
kmeans.fit(X)
```

```python
np.unique(kmeans.labels_)
```

```python
df['labels']
```

```python
label_mapping = {1: 0, 2: 1, 3: 2, 4: 3, 5: 4}
df['new_labels'] = df['labels'].map(label_mapping)
```

```python
silhouette_score(X, kmeans.labels_)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(df['new_labels'],kmeans.labels_ ))
```
