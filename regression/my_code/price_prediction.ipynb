```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
```

```python
dataset=pd.read_csv(r'C:\Users\mahdi\Documents\jupyter-notebook\machine_learning_source\Car details.csv')
dataset.head()
```

```python
df=dataset.copy()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df[['mileage_number','mileage unit']]=df['mileage'].str.split(' ',  expand=True)
df.drop(['mileage', 'mileage unit'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['mileage unit'].unique()
```

```python
df[['engine_number','engine_unit']]=df['engine'].str.split(' ',  expand=True)
df.drop(['engine', 'engine_unit'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df[['max_power_number','max_power_unit']]=df['max_power'].str.split(' ',  expand=True)
df.drop(['max_power', 'max_power_unit'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df[['torque_number']]=df['torque'].str.split('@',  expand=True).drop([1,2], axis=1)
```

```python
df['torque_number']=df['torque_number'].str.extract('(^\d*)')
```

```python
df['torque_number']=df['torque_number'].astype(float)
```

```python
def change_unit(x):
    if x <= 50:
        return x * 9.8 
    else:
        x
```

```python
df['torque_number']=df['torque_number'].apply(change_unit)
```

```python
df.head()
```

```python
df['mileage_number']=df['mileage_number'].astype(float)
df['mileage_number'].fillna(df['mileage_number'].mean(), inplace=True)
df['engine_number']=df['engine_number'].astype(float)
df['engine_number'].fillna(df['engine_number'].mean(), inplace=True)
df['max_power_number'].replace({'':np.nan}, inplace=True)
df['max_power_number']=df['max_power_number'].astype(float)
df['max_power_number'].fillna(df['max_power_number'].mean(), inplace=True)
df['torque_number'].fillna(df['torque_number'].mean(), inplace=True)
df['seats'].fillna(df['seats'].mean(), inplace=True)
```

```python
df.info()
```

```python
df=df.drop('name', axis=1)
```

```python
col_of_categorical=[]
for col in df.columns:
    if df[col].dtype=='object':
        col_of_categorical.append(col)
col_of_categorical
```

```python
df=pd.concat([df, pd.get_dummies(df[col_of_categorical], dtype=int, drop_first=True)], axis=1 )
df.drop(df[col_of_categorical], axis=1, inplace=True)
```

```python
from datetime import date
df['age_of_car']=date.today().year-df['year']
df=df.drop('year', axis=1)
```

```python
df.head()
```

```python
X = df.iloc[:, 1:]
y = df.iloc[:, 0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
regressor_lr=LinearRegression()
regressor_lr.fit(X_train, y_train)
```

```python
regressor_lr.score(X_train, y_train)
```

```python
regressor_lr.score(X_test, y_test)
```

```python
pred_lr=regressor_lr.predict(X_test)
```

```python
plt.hist(pred_lr-y_test, bins=20)
plt.show()
```

```python
plt.scatter(pred_lr, y_test)
plt.show()
```

```python
regressor_rf=RandomForestRegressor()
regressor_rf.fit(X_train, y_train)
```

```python
regressor_rf.score(X_train, y_train)
```

```python
regressor_rf.score(X_test, y_test)
```

```python
sc2=StandardScaler()
y_train=sc2.fit_transform(pd.DataFrame(y_train))
y_test=sc2.transform(pd.DataFrame(y_test))
```

```python
regressor_svr=SVR()
regressor_svr.fit(X_train, y_train)
```

```python
regressor_svr.score(X_train, y_train)
```

```python
regressor_svr.score(X_test, y_test)
```

```python
pred_svr = regressor_svr.predict(X_test)
```

```python
plt.hist(pred_svr.reshape(-1,1)-y_test, bins=20)
plt.show()
```

```python
plt.scatter(pred_svr, y_test)
plt.show()
```

```python
from sklearn import metrics
print('MAE = ', metrics.mean_absolute_error(sc2.inverse_transform(y_test), sc2.inverse_transform(pred_svr.reshape(-1,1))))
print('MSE = ', metrics.mean_squared_error(sc2.inverse_transform(y_test), sc2.inverse_transform(pred_svr.reshape(-1,1))))
print('RMSE = ', np.sqrt(metrics.mean_squared_error(sc2.inverse_transform(y_test), sc2.inverse_transform(pred_svr.reshape(-1,1)))))
print('R2 = ', metrics.r2_score(sc2.inverse_transform(y_test), sc2.inverse_transform(pred_svr.reshape(-1,1))))
```
