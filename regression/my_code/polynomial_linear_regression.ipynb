```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'C:\Users\mahdi\Documents\jupyter-notebook\machine_learning_source\Position_Salaries.csv')
df
```

```python
X=df.iloc[:, 1:-1].values
y=df.iloc[:, -1].values
```

```python
y
```

```python
plt.scatter(X, y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
```

```python
regressor.fit(X, y)
```

```python
plt.scatter(X, y)
plt.plot(X, regressor.predict(X), color='red')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=4)
```

```python
X_poly=poly.fit_transform(X)
X_poly
```

```python
regressor2=LinearRegression()
```

```python
regressor2.fit(X_poly, y)
```

```python
plt.scatter(X,y)
plt.plot(X, regressor2.predict(X_poly), color='red')
plt.show()
```

```python
regressor2.predict(poly.fit_transform([[6.5]]))
```
